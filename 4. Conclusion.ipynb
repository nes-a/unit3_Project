## 1.  Load the data:

```python
import pandas as pd
```

```python
df = pd.read_csv('test_summary.csv')
df = df.drop('Unnamed: 0', axis=1)
```

```python
df.head()
```

## 2. Analysis:

```python
unique_clusters = sorted(df['cluster'].unique())
print(f"\nFound {len(unique_clusters)} unique clusters: {unique_clusters}")

# Iterate through each cluster
for cluster_id in unique_clusters:
    cluster_df = df[df['cluster'] == cluster_id]
    total_samples_in_cluster = len(cluster_df)

    if total_samples_in_cluster == 0:
        print(f"\nCluster N°{cluster_id}: No samples found.")
        continue

    print(f"\n--- Cluster N°{cluster_id} (Total samples: {total_samples_in_cluster}) ---")

    actual_category_counts = cluster_df['Category'].value_counts(normalize=True) * 100
    predicted_class_counts = cluster_df['Predicted_class'].value_counts(normalize=True) * 100

    cluster_analysis_results = {}
    all_classes = sorted(list(df['Category'].unique()) + list(df['Predicted_class'].unique()))

    for cls in all_classes:
        cluster_analysis_results[f'true_class{cls}_percentage'] = actual_category_counts.get(cls, 0.0)
        cluster_analysis_results[f'predicted_class{cls}_percentage'] = predicted_class_counts.get(cls, 0.0)

    print(f"Cluster N°{cluster_id} Analysis:")
    for key, value in cluster_analysis_results.items():
        print(f"  {key}: {value:.2f}%")
```

## 3. Explanations and insights :

```python
print("Based on the percentages above, we can observe the following:")
print("""- For Cluster N°0, shows perfect alignment, with both true and predicted classes being 100% class4.
This small cluster was perfectly classified.""")
print("""- Clusters N°1 (88 samples) & N°2 (26 samples): Predominantly true_class0, and the model largely predicts
class0 accurately within these. However, the model misclassifies minority true classes (e.g., true_class1,
true_class2, true_class3) within these clusters, often predicting them as class0.""")
print("""- Cluster N°4 (6 samples): This is a highly mixed cluster with diverse true classes. The model struggles
significanty here, misclassifying most samples and predicting only class0 and class4, indicating that the features
defining this cluster are not distinct enough for accurate classification of its varied true classes.""")
```
